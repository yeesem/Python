```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
#Multiplot
#Number of rows,number of col,numbering
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented

#Create a figure
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
#Subplot by using Object Oriented Method
fig,axes = plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)

for current_ax in axes:
   current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

#Avoid graphs overlapping
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2,figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#label is for the reference of legend function
ax.plot(x,x**2,label = 'X Square')
ax.plot(x,x**3,label = 'X Cubed')

#loc the location of the legend box
#loc=(0.1,0.1)
#ax.legend(loc = 0)
ax.legend(loc = (0.1,0.1))


ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#Can set the colour by using the RGB Hex code for more options
#alpha - set the transparency of the line
#Type lw also works to replace linewidth
#ls = linestyle
ax.plot(x,y,color = 'blue',linewidth=2,alpha = 0.5,linestyle=':',marker = 'o',markersize = 5
        ,markerfacecolor = 'yellow',markeredgewidth = 2,markeredgecolor = 'black')
```

```python
fig =plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'blue',lw=2,alpha = 0.5,ls = '--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
