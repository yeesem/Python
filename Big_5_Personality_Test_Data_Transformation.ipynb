```python
extro_ques = [ i for i in range(1,45,5) if i!= 41]
aggre_ques = [i for i in range(2,45,5)]
cons_ques = [i for i in range(3,45,5)]
neuro_ques = [i for i in range(4,45,5) if i!= 44]
open_ques = [i for i in range(5,45,5)]

open_ques.append(41)
open_ques.append(44)
```

```python
print(extro_ques_index)
print(aggre_ques_index)
print(cons_ques_index)
print(neuro_ques_index)
print(open_ques_index)
```

```python
import numpy as np
x = 0
extro_mark = []
for i in range(len(df2)):
  for j in extro_ques:
    x += df2.iloc[i][f"Q{j}"]
  extro_mark.append(x)
  x = 0
```

```python
print(extro_mark)
len(extro_mark)
```

```python
df2['Extroversion Mark'] = [df2.iloc[i,extro_ques_index].sum() for i in range(len(df2))]
df2['Agreeableness Mark'] = [df2.iloc[i,aggre_ques_index].sum() for i in range(len(df2))]
df2['Conscientiousness Mark'] = [df2.iloc[i,cons_ques_index].sum() for i in range(len(df2))]
df2['Neuroticism Mark'] = [df2.iloc[i,neuro_ques_index].sum() for i in range(len(df2))]
df2['Openness Mark'] = [df2.iloc[i,open_ques_index].sum() for i in range(len(df2))]
```

```python
df2.head()
```

```python
df2["Extroversion"] = df2["Extroversion Mark"].apply(lambda x : 1 if (x/(len(extro_ques)))>3 else 0)
df2["Agreeableness"] = df2["Agreeableness Mark"].apply(lambda x : 1 if (x/(len(aggre_ques)))>3 else 0)
df2["Conscientiousness"] = df2["Conscientiousness Mark"].apply(lambda x : 1 if (x/(len(cons_ques)))>3 else 0)
df2["Neuroticism"] = df2["Neuroticism Mark"].apply(lambda x : 1 if (x/(len(neuro_ques)))>3 else 0)
df2["Openness"] = df2["Openness Mark"].apply(lambda x : 1 if (x/(len(open_ques)))>3 else 0)
```
