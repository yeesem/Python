```python
def calculate_dividend(dividend):
  hashmap = {}
  for i in range(len(dividend)):
     dividend_index = str(dividend.index[i])
     if hashmap.get(dividend_index[:4]) is not None:
      hashmap[dividend_index[:4]] = hashmap.get(dividend_index[:4]) + dividend[dividend_index]
     else:
      list = []
      hashmap[dividend_index[:4]] = dividend[dividend_index]

  return hashmap
```
